# Taylor series graphical example

This notebook shows an example of Taylor series approximations in Julia. We will use the sine function evlauated in $\pi/2$ as the starting point and we will show how the different approximations fit the initial curves. 

$$f(x) = sin(x)$$

**Note:** This notebook also works fine for introducing basic Julia concepts as creating arrays, plotting and using `for` loops. 

```julia
# Library for plotting
using Plots

# Defining points where the approximation is going to be shown
x0 = pi / 2
x1 = 3 * pi / 2
fx0 = sin(x0)

# Defining space to plot sine function
x = LinRange(x0, x1, 200)

# Plotting the function just to be sure that e're making the right thing
plot(x, sin.(x), label="y = sin(x)", gridlinewidth = 1.5)
xlabel!("x")
ylabel!("sin(x)")
ylims!((-1, 1))
xlims!((x0, x1))
title!("Sine fuction")
```

Defining the values of the first four derivatives of the sine function: 
 
$f(x) = sen(x) \rightarrow f(x_0) = 1 \rightarrow f(x_1) = -1$

$f'(x) = cos(x) \rightarrow f'(x_0)=0 \rightarrow f'(x_1)=0$

$f''(x) = -sen(x) \rightarrow f''(x_0)=-1 \rightarrow f''(x_1)=1$

$f'''(x) = -cos(x) \rightarrow f'''(x_0) = 0 \rightarrow f'''(x_1) = 0$

$f''''(x)= sen(x) \rightarrow f''''(x_0)=1 \rightarrow f''''(x_1) = -1$

Given the results, we can propose the following infinite series that wil approximate the sine function. 

$$f(x) = \sum_{n=0}^{\infty} (-1)^n\frac{(x - \pi/2)^{2n}}{(2n)!}$$

```julia
# Defining the Taylor series according to the expansion
terms = 6

# Defining the vector that will hold the results
y_app = zeros(length(x))

# Plotting the analytical function to see the approximations
p = plot(   x, sin.(x), 
        label = "Analytical",
        lw=2.5, 
        seriestype =:path, 
        linestyle =:dash,
        gridlinewidth=1.5)
ylims!((-1, 1))
xlims!((x0, x1))

# Defining loop to calculate and plot
for n in 0:terms

    # Calculating the approximation for the current order
    y_app = y_app .+ (-1)^n * (x .- x0).^(2 * n) / (factorial(2 * n))

    # Creating plot id for legend
    seal = "Approx. order =  " * string(2 * n)

    # Plotting the solution
    plot!(x, y_app, label=seal)


end

display(p)
```
