# INTEGRACIÓN NUMÉRICA

Evaluar integrales es un quehacer muy común en la práctica de la ingeniería. Para poder hacerlo, se puede aplicar diferentes tácticas que dan diferentes resultados en cuanto se refiere al nivel de precisión de la integral. Básicamente, este notebook explora dos formas de integración con diferentes resultados: 

1. Fórmulas de Newton-Cotes
2. Cuadratura de Gauss

Estos métodos se utlizan para integrales definidas en intevalos $a-b$. Para integraciones con límites abiertos, estos métodos se pueden extender, pero eso se desarrollará en otro notebook. 

Para poder desarrollar el ejercicio con un ejemplo, se resolverá la integral: 

$$\int_0^{\frac{\pi}{2}}cos(x)dx$$

Se puede probar fácilmente que la solución analítica de la integral es 1.0

$$\int_0^{\frac{\pi}{2}}cos(x)dx = sen(x)\Big|_0^{\frac{\pi}{2}} = 1 - 0 = 1$$

## Aproximación "naive" (regla del trapecio)

Como primera aproximación, podría definir una interpolación lineal entre los límites de la integral y calcular el área bajo esta línea recta. Este método se llama *regla del trapecio* y no tiene resultados muy convincentes, especialmente si la función a integrar es una curva compleja y si los límites de la integral están relativamente lejos. 

```python
# Definir un vector x para poder graficar
x = np.linspace(0, np.pi / 2, 300)
y = np.cos(x)

# Graficando la función y la aproximación de la integral
plt.plot(x, y, ls = '--', label=r'$f(x)=cos(x)$')
plt.plot([x[0], x[-1]], [y[0], y[-1]], label='Función aproximada')
plt.fill_between([x[0], x[-1]], 0, [y[0], y[-1]], color='orange', alpha = 0.4,
                 label='Integral calculada')
plt.xlim([0, np.pi / 2])
plt.ylim([0, 1])
plt.grid()
plt.legend()
plt.title(r'Integral de $f(x) = cos(x)$ con regla del trapecio')
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.tight_layout()
plt.show()
```

La regla del trapecio sencillo calcula la integral entre los límites $a$ y $b$ se resume en la siguiente expresión:

$$I \approx \frac{(b-a)}{2}\big[f(a) + f(b)\big]$$
